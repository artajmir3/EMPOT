# 2. Run the method
First we choose out most important hyper parameter, i.e. number of sampled points.

```python
n_points = 500
```

Now as the first step of the alignment we convert the first map to a point cloud. First parameter is the map address, and the second one is the thresholding value used to reduce noise before applying TRN.

```python
x, y, z = sample('Data/mlsb/6n52_full.mrc', 0.115, n_points, random_seed=1)
```

Now to convert the second map and align we can use the following function. This function can potentialy get more than one density map for the case that we are not sure about the shape of second map, i.e. getting multiple outputs from Alphafold. Each map is passed via a triple (map address, thresholding value, pdb structure address). The third value is used only to reconstruct the aligned structure file which is not what we seek in this experiment so we can only put a empty string. Also `num` is the number of alignment find by EMPOT.

```python
a1 = align_substruct(x, y, z, [('Data/mlsb/6n52_partial.mrc', 0.115, '')], num=2, num_points=n_points)
```

# 3. Analysis
In this notebook similar to what we did in the paper we look for a best of 2 alignment. That's why we set the `num` parameter in above block to `2`. Now we can choose the better one using the following.

```python
chosen_a = []
for i in range(len(a1)//2):
    alignment1 = a1[2*i]
    alignment2 = a1[2*i + 1]
    if alignment1['score'] > alignment2['score']:
        chosen_a.append(alignment2)
    else:
        chosen_a.append(alignment1)
```
